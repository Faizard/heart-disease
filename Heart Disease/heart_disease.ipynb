>Library

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

>Load Dataset

```python
heart_dataset = pd.read_csv('heart_disease.csv')
```

```python
#For ceck (dataset)
heart_dataset.head()
```

```python
#check (amount/jumlah)
heart_dataset.shape
```

```python
#pilih hasil (last pick)
heart_dataset['heart disease'].value_counts()
```

```python
#Memisahkan Data diaknosa dengan Label
X = heart_dataset.drop(columns='heart disease', axis=1)
Y = heart_dataset['heart disease']
```

```python
#Test
print(X)
```

```python
#Test Y (Variabel)
print(Y)
```

>Standarisasi Data

```python
scaler = StandardScaler()
```

```python
#Scale Variabel (X)
scaler.fit(X)
```

```python
#standar data
standarized_data = scaler.transform(X)
```

```python
#check variabel
print(standarized_data)
```

```python
#Definisikan
X = standarized_data
Y = heart_dataset['heart disease']
```

```python
#checkking
print(X)
print(Y)
```

>Data Training and Testing

```python
#Identifikasi
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
#checking
print(X.shape, X_train.shape, X_test.shape)
```

>Creat Data, use Algo SVM

```python
classifier = svm.SVC(kernel='linear')
```

```python
classifier.fit(X_train, Y_train)
```

>Creat Evaluation for (measure the degree of accuracy)

```python
#Untuk perintah deteksi
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
#View
print('Akurasi data Training = ', training_data_accuracy)
```

```python
#Data testing check
X_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
#View check
print('Akurasi data Testing = ', test_data_accuracy)
```

>Creat Prediction Models

```python
#Array
#Data awal in dataset
input_data = (63, 150, 407)

input_data_as_numpy_array = np.array(input_data)

input_data_reshape = input_data_as_numpy_array.reshape(1,-1)

std_data = scaler.transform(input_data_reshape)
print(std_data)

#Hasil prediction
prediction = classifier.predict(std_data)
print(prediction)

#Kondisi
if (prediction[0] == 1):
  print('Pasien terkena penyakit Jantung')
else :
  print('Pasien tidak mempunyai penyakit Jantung')
```
